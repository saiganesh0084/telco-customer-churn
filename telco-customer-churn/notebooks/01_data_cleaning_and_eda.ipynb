### Data Type Issue: TotalCharges

The `TotalCharges` column is stored as an object instead of numeric.
This can break statistical analysis and machine learning models.


```python
df['TotalCharges'].unique()[:10]
```

```python
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df[df['TotalCharges'].isnull()][['tenure', 'MonthlyCharges', 'TotalCharges']].head()
```

```python
# Option 1: Assign back explicitly
df['TotalCharges'] = df['TotalCharges'].fillna(0)

# Option 2: Use DataFrame-level fillna
df.fillna({'TotalCharges': 0}, inplace=True)
```

### Missing Value Handling

After converting `TotalCharges` to numeric, missing values appeared.
These rows correspond to customers with zero tenure, so filling with 0
reflects correct business logic and avoids false revenue.


```python
df[df['TotalCharges'].isnull()][['tenure', 'MonthlyCharges', 'TotalCharges']].head()
```

```python
num_cols = ['tenure', 'MonthlyCharges', 'TotalCharges']
df[num_cols].describe()
```

```python
Q1 = df['MonthlyCharges'].quantile(0.25)
Q3 = df['MonthlyCharges'].quantile(0.75)
IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

lower, upper
```

```python
df[(df['MonthlyCharges'] < lower) | (df['MonthlyCharges'] > upper)].head()
```
